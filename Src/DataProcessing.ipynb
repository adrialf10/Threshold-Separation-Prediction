```python
# Dataset con la información de separación
r32_df = pd.read_csv('Runway32.csv', header = 0, decimal = ',', sep = ';')

r32_df = r32_df.drop(['date', 'dateTime', 'next_dateTime', 'arr_volume_group', 'runway', 'flight_key', 'metarKey', 'maxKnots'], axis=1)

r32_df.loc[(r32_df['m4_altitude'] == '27+'), 'm4_altitude'] = 27
r32_df = r32_df.astype({'m4_altitude': 'float64'})

enc = OrdinalEncoder(categories = [['L', 'M', 'H', 'J']])
r32_df.loc[:, ['flight_wake']] = enc.fit_transform(r32_df[['flight_wake']])

enc = OrdinalEncoder(categories = [['Front', 'DFront', 'DTail', 'Lateral', 'Tail']])
r32_df.loc[:, ['direction']] = enc.fit_transform(r32_df[['direction']])

r32_df
```

```python
# Dataset con la información de separación
r18_df = pd.read_csv('Runway18.csv', header = 0, decimal = ',', sep = ';')

r18_df = r18_df.drop(['date', 'dateTime', 'next_dateTime', 'arr_volume_group', 'runway', 'flight_key', 'metarKey', 'maxKnots'], axis=1)

r18_df.loc[(r18_df['m4_altitude'] == '27+'), 'm4_altitude'] = 27
r18_df = r18_df.astype({'m4_altitude': 'float64'})

enc = OrdinalEncoder(categories = [['L', 'M', 'H', 'J']])
r18_df.loc[:, ['flight_wake']] = enc.fit_transform(r18_df[['flight_wake']])

enc = OrdinalEncoder(categories = [['Front', 'DFront', 'DTail', 'Lateral', 'Tail']])
r18_df.loc[:, ['direction']] = enc.fit_transform(r18_df[['direction']])

#r18_df
```

```python
pd.DataFrame(r32_df.var())
```

```python
X = r32_df.drop(['threshold_separation'], axis=1)
y = r32_df['threshold_separation']
```

```python
# Estandarizacion
a = StandardScaler().fit_transform(X)
ss_df = pd.DataFrame(data = a, columns = X.columns)
```

```python
# PCA Estandarizado
pca = PCA(n_components = 10)
pca_a = pca.fit_transform(ss_df)
pca_df = pd.DataFrame(data = pca_a, columns = ['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10'])
#pca_df = pd.DataFrame(data = pca_a)
```

```python
# Generacion test-train
X_train, X_test, y_train, y_test = train_test_split(ss_df, y, shuffle=True, train_size = 0.7)

# Random Forest Regression
model = RandomForestRegressor(n_estimators = 100)
model.fit(X_train, y_train)

prediction = model.predict(X_test)

pd.DataFrame(data = model.feature_importances_, index = X_train.columns)
```

```python
ss_df = pd.merge(y, ss_df, left_index=True, right_index=True)
pca_df = pd.merge(y, pca_df, left_index=True, right_index=True)
```

```python
ss_df.to_csv('StandarizedRunway32.csv', decimal = ',', sep = ';', index = False)
pca_df.to_csv('PCARunway32.csv', decimal = ',', sep = ';', index = False)
```
